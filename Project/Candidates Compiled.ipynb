```python
# Find hashtags
import re

def find_hashtag(hasht):
    a_list = re.findall(r"#(\w+)", hasht)
    new_a_list = [x.lower() for x in a_list]
    return new_a_list

df['hashtags'] = df['tweet_text'].apply(find_hashtag)
```

```python
# Find hashtags with warren
def find_warren(hasht):
    for name in hasht: 
        found = re.findall(r"(elizabethwarren)", name)
        if found:
            for f in found: 
                return str(f)
        

df['warren'] = df['hashtags'].apply(find_warren)
```

```python
# Find hashtags with pete
def find_pete(hasht):
    for name in hasht: 
        found = re.findall(r"(petebuttigieg)", name)
        if found:
            for f in found: 
                return str(f)
        

df['pete'] = df['hashtags'].apply(find_pete)

# Find hashtags with bernie

def find_bernie(hasht):
    for name in hasht: 
        found = re.findall(r"(berniesanders)", name)
        if found:
            for f in found: 
                return str(f)
        
df['berne'] = df['hashtags'].apply(find_bernie)

def find_biden(hasht):
    for name in hasht: 
        found = re.findall(r"(joebiden)", name)
        if found:
            for f in found: 
                return str(f)
        

df['biden'] = df['hashtags'].apply(find_biden)

def find_bloomberg(hasht):
    for name in hasht: 
        found = re.findall(r"(bloomberg)", name)
        if found:
            for f in found: 
                return str(f)
        

df['bloomberg'] = df['hashtags'].apply(find_bloomberg)
# df['bloomberg']

def find_trump(hasht):
    for name in hasht: 
        found = re.findall(r"(trump)", name)
        if found:
            for f in found: 
                return str(f)
        

df['trump'] = df['hashtags'].apply(find_trump)
# df['bloomberg']
```

```python
# Create politican dataframe, choose the columns to include and rename the columns to politician

warren = df[df['warren']=='elizabethwarren'][['username', 'warren']].rename(columns={"warren": "politician"})


pete = df[df['pete']=='petebuttigieg'][['username', 'pete']].rename(columns={"pete": "politician"})

bernie = df[df['berne']=='berniesanders'][['username', 'berne']].rename(columns={"berne": "politician"})

biden = df[df['biden']=='joebiden'][['username', 'biden']].rename(columns={"biden": "politician"})

bloom = df[df['bloomberg']=='bloomberg'][['username', 'bloomberg']].rename(columns={"bloomberg": "politician"})
trump = df[df['trump']=='trump'][['username', 'trump']].rename(columns={"trump": "politician"})

df_total = warren.append([pete, bernie, biden, bloom,trump])
```

```python
df_total.drop_duplicates(inplace=True)
```

```python
df_total = df_total[df_total['username'] != 'AAA']
```
