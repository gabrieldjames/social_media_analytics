```python
biden = pd.read_csv('biden_total.csv')
bernie = pd.read_csv('feel_the_bern.csv')
trump = pd.read_csv('trumptweetdata_total.csv').dropna()
pete = pd.read_csv('PeteButtigieg.csv')
war = pd.read_csv('warren_combined_twitter.csv')
mike1 = pd.read_csv('bloomberg_Twitter_data.csv')
mike2 = pd.read_csv('bloomberg_Twitter_data2.csv')
mike3 = pd.read_csv('bloomberg_Twitter_data3.csv')
mike = mike1.append(mike2)
mike = mike.append(mike3).dropna()
```

```python
# Find hashtags
import re

def find_hashtag(hasht):
    a_list = re.findall(r"#(\w+)", hasht)
    new_a_list = [x.lower() for x in a_list]
    return new_a_list

df['hashtags'] = df['tweet_text'].apply(find_hashtag)
```

```python
# Find hashtags with warren
def find_warren(hasht):
    for name in hasht: 
        found = re.findall(r"(warren)", name)
        if found:
            for f in found: 
                return str(f)
        

df['warren'] = df['hashtags'].apply(find_warren)
```

```python
# Find hashtags with pete
def find_pete(hasht):
    for name in hasht: 
        found = re.findall(r"(petebuttigieg)", name)
        if found:
            for f in found: 
                return str(f)
        

df['pete'] = df['hashtags'].apply(find_pete)

# Find hashtags with bernie

def find_bernie(hasht):
    for name in hasht: 
        found = re.findall(r"(berniesanders)", name)
        if found:
            for f in found: 
                return str(f)
        
df['berne'] = df['hashtags'].apply(find_bernie)

def find_biden(hasht):
    for name in hasht: 
        found = re.findall(r"(joebiden)", name)
        if found:
            for f in found: 
                return str(f)
        

df['biden'] = df['hashtags'].apply(find_biden)

def find_bloomberg(hasht):
    for name in hasht: 
        found = re.findall(r"(bloomberg)", name)
        if found:
            for f in found: 
                return str(f)
        

df['bloomberg'] = df['hashtags'].apply(find_bloomberg)
# df['bloomberg']

def find_trump(hasht):
    for name in hasht: 
        found = re.findall(r"(trump)", name)
        if found:
            for f in found: 
                return str(f)
        

df['trump'] = df['hashtags'].apply(find_trump)
# df['bloomberg']
```

```python
# Create politican dataframe, choose the columns to include and rename the columns to politician

warren = df[df['warren']=='warren'][['username', 'warren']].rename(columns={"warren": "politician"})


pete = df[df['pete']=='petebuttigieg'][['username', 'pete']].rename(columns={"pete": "politician"})

bernie = df[df['berne']=='berniesanders'][['username', 'berne']].rename(columns={"berne": "politician"})

biden = df[df['biden']=='joebiden'][['username', 'biden']].rename(columns={"biden": "politician"})

bloom = df[df['bloomberg']=='bloomberg'][['username', 'bloomberg']].rename(columns={"bloomberg": "politician"})
trump = df[df['trump']=='trump'][['username', 'trump']].rename(columns={"trump": "politician"})

df_total = warren.append([pete, bernie, biden, bloom,trump])
```

# NetworkX

Bernie

```python
df_bernie = df[df['berne'] == 'berniesanders']
name_list = df_bernie['original_creator'].value_counts().index.values[:200]

def name_in_list(username): 
    if username in name_list: 
        return 1 
    return 0 

df_bernie['degrees'] = df_bernie['original_creator'].map(name_in_list)
df_bernie = df_bernie[df_bernie['degrees'] == 1]



# Create retweet network from edgelist
network = nx.from_pandas_edgelist(
    df_bernie,
    source = 'username', 
    target = 'original_creator',
    create_using = nx.DiGraph())

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 15))
nx.draw(network, edge_color ="black", linewidths=0.3, node_size=50, node_color = 'blue')
plt.title('Pete Buttigieg', size=15)
plt.show()
```
