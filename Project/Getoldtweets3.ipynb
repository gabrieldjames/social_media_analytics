```python
import GetOldTweets3 as got
from pandas import DataFrame,Series
import pandas as pd
```

```python
x = 0 
tweet_username = []
tweets= []
date = []
rt = []

tweetCriteria = got.manager.TweetCriteria().setQuerySearch('#berniesanders').setSince("2019-08-01").setUntil("2020-02-19").setMaxTweets(2000)

```

```python

for x in range(1999):
    print(x)
    twe = got.manager.TweetManager.getTweets(tweetCriteria)[x]
    tweet_username.append(twe.username)
    tweets.append(twe.text)
    date.append(twe.date)
    rt.append(twe.retweets)
```

```python
tweet_df = DataFrame()
tweet_df['user'] = tweet_username
tweet_df['tweet'] = tweets
tweet_df['rt'] = retweets
tweet_df['date'] = date
```
